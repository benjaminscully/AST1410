```python
import mesa_reader as mr
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use('/Users/scully/Documents/mplstyles/paper.mplstyle')
```

```python
def make_plots(pre_ms_file, ms_file, post_ms_file, alpha):
    h_pre_ms = mr.MesaData(pre_ms_file)
    h_ms = mr.MesaData(ms_file)
    h_post_ms = mr.MesaData(post_ms_file)

    log_Teff_pre_ms = h_pre_ms.log_Teff
    log_L_pre_ms = h_pre_ms.log_L
    log_P_c_pre_ms = h_pre_ms.log_cntr_P
    log_rho_c_pre_ms = h_pre_ms.log_cntr_Rho
    log_T_c_pre_ms = h_pre_ms.log_cntr_T

    log_Teff_ms = h_ms.log_Teff
    log_L_ms = h_ms.log_L
    log_P_c_ms = h_ms.log_cntr_P
    log_rho_c_ms = h_ms.log_cntr_Rho
    log_T_c_ms = h_ms.log_cntr_T

    log_Teff_post_ms = h_post_ms.log_Teff
    log_L_post_ms = h_post_ms.log_L
    log_P_c_post_ms = h_post_ms.log_cntr_P
    log_rho_c_post_ms = h_post_ms.log_cntr_Rho
    log_T_c_post_ms = h_post_ms.log_cntr_T

    # make a 1x3 plot with each of the above plots in a row to be put in a LateX doc
    fig, axs = plt.subplots(1, 3, figsize=(15, 5))

    # plot HR diagram
    axs[0].plot(log_Teff_ms, log_L_ms, label="Main Sequence")
    axs[0].plot(log_Teff_pre_ms, log_L_pre_ms, label="Pre Main Sequence")
    axs[0].plot(log_Teff_post_ms, log_L_post_ms, label="Post Main Sequence")
    axs[0].set_xlabel(r"$\log(T_{\rm eff})$")
    axs[0].set_ylabel(r"$\log(L)$")
    axs[0].set_title(r"H-R Diagram for 15 $M_\odot$ star with $\alpha={:.1f}$".format(alpha))
    axs[0].invert_xaxis()
    axs[0].legend()

    # plot central density vs central pressure
    axs[1].plot(log_P_c_ms, log_rho_c_ms, label="Main Sequence")
    axs[1].plot(log_P_c_pre_ms, log_rho_c_pre_ms, label="Pre Main Sequence")
    axs[1].plot(log_P_c_post_ms, log_rho_c_post_ms, label="Post Main Sequence")
    axs[1].set_xlabel(r"$\log(P_c)$")
    axs[1].set_ylabel(r"$\log(\rho_c)$")
    axs[1].set_title(r"Central Pressure vs Central Density with $\alpha={:.1f}$".format(alpha))
    # axs[1].set_xlim(15, 25)

    # plot central temperature vs central density
    axs[2].plot(log_T_c_ms, log_rho_c_ms, label="Main Sequence")
    axs[2].plot(log_T_c_pre_ms, log_rho_c_pre_ms, label="Pre Main Sequence")
    axs[2].plot(log_T_c_post_ms, log_rho_c_post_ms, label="Post Main Sequence")
    axs[2].set_xlabel(r"$\log(T_c)$")
    axs[2].set_ylabel(r"$\log(\rho_c)$")
    axs[2].set_title(r"Central Temperature vs Central Density with $\alpha={:.1f}$".format(alpha))
    # axs[2].set_xlim(7, 9.5)

    plt.tight_layout()
    plt.savefig("a{}_combined_plots.png".format(alpha), bbox_inches='tight', dpi=300)
    plt.show()
```

```python
# Make plots for different alpha values

# ==== Question 1 ====
# alpha = 2
make_plots("./p1_data/pre_ms.data", "./p1_data/ms.data", "./p1_data/post_ms.data", 2)

# ==== Question 2 ====
# alpha = 1
make_plots("./alpha_1.0_data/pre_ms.data", "./alpha_1.0_data/ms.data", "./alpha_1.0_data/post_ms.data", 1)

# alpha = 4
make_plots("./alpha_4.0_data/pre_ms.data", "./alpha_4.0_data/ms.data", "./alpha_4.0_data/post_ms.data", 4)
```
