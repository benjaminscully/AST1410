```python
import numpy as np
import astropy.units as u
import astropy.constants as const
```

```python
m_p = 1.007852 * u.u
m_He4 = 1.000651 * u.u
M_sun = const.M_sun.cgs
L_sun = const.L_sun.cgs
c = const.c.cgs
```

```python
eps_MS = (1 - m_He4 / m_p) * (1 - 0.022)
print("eps_MS:", eps_MS)
print("Energy released (MeV):", ((4*m_p)*eps_MS*c**2).to(u.MeV))

E_MS = eps_MS * 0.1 * M_sun * c**2
print("E_MS (erg):", E_MS.to(u.erg))
tau_MS = E_MS / L_sun
print("tau_MS:", tau_MS.to(u.Gyr))

mu_MS = 1 * u.u
mu_PMS = 1.3 * u.u

print("mu_MS:", mu_MS)
print("mu_PMS:", mu_PMS)

M_f = M_sun - E_MS / c**2
print("M_f:", M_f.to(u.M_sun))

tau_PMS = tau_MS * (M_f / M_sun)**(-58/13) * (mu_PMS / mu_MS)**(-114/13)
print("tau_PMS:", tau_PMS.to(u.Gyr))
```
