```python
import mesa_reader as mr
import numpy as np
import matplotlib.pyplot as plt
```

```python
h_pre_ms = mr.MesaData("pre_ms.data")
h_ms = mr.MesaData("ms.data")
h_post_ms = mr.MesaData("post_ms.data")

log_Teff_pre_ms = h_pre_ms.log_Teff
log_L_pre_ms = h_pre_ms.log_L

log_Teff_ms = h_ms.log_Teff
log_Teff_post_ms = h_post_ms.log_Teff

log_L_ms = h_ms.log_L
log_L_post_ms = h_post_ms.log_L

log_Teff = np.concatenate((log_Teff_pre_ms, log_Teff_ms, log_Teff_post_ms))
log_L = np.concatenate((log_L_pre_ms, log_L_ms, log_L_post_ms))
```

```python
plt.plot(log_Teff_ms, log_L_ms, label="Main Sequence")
plt.plot(log_Teff_pre_ms, log_L_pre_ms, label="Pre Main Sequence")
plt.plot(log_Teff_post_ms, log_L_post_ms, label="Post Main Sequence")
# plt.plot(log_Teff, log_L)
plt.xlabel("log(Teff)")
plt.ylabel("log(L)")
plt.title("Evolutionary track of 15 solar mass star on H-R Diagram")
plt.gca().invert_xaxis()
plt.legend()
plt.savefig("hr_diagram.png", bbox_inches='tight', dpi=300)
plt.show()
```
