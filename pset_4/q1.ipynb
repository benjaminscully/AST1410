```python
import numpy as np
import astropy.units as u
import astropy.constants as const
```

```python
R_gps = 4.2164e9 * u.cm
R_earth = 6.371e8 * u.cm
M_earth = 5.972e27 * u.g

sidereal_day = (1 * u.sday).to(u.s)
G = const.G.cgs
c = const.c.cgs

U_gps = G * M_earth / R_gps
U_earth = G * M_earth / R_earth

v_gps = 2 * np.pi * R_gps / sidereal_day
v_earth = 2 * np.pi * R_earth / sidereal_day

t_gps_over_t_earth = (1 - (0.5 * v_gps**2 + U_gps) / c**2) / (1 - (0.5 * v_earth**2  + U_earth) / c**2) 
print(t_gps_over_t_earth)
print(t_gps_over_t_earth - 1)
```

```python
print(sidereal_day.to(u.s))
print((sidereal_day * (t_gps_over_t_earth - 1)).to(u.us))
```
