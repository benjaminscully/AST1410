```python
import numpy as np
import astropy.units as u
import astropy.constants as const
```

```python
P_c1 = 1e17 * u.g * u.cm**-1 * u.s**-2
rho_c1 = 60 * u.g * u.cm**-3
M_1 = const.M_sun.cgs

m_e = const.m_e.cgs
m_u = const.u.cgs
h = const.h.cgs
K_NR = h**2 / (20 * m_e * (m_u)**(5/3)) * (3 / np.pi)**(2/3)
K_NR = K_NR.to(u.cm**4 / u.s**2 / u.g**(2/3))
print(f"{K_NR:.4e}")

T = 1e8 * u.K
k_B = const.k_B.cgs

mu_e = 2
mu = 4/3
```

```python
rho_2 = ((T*k_B*mu_e**(5/3))/(K_NR * mu * m_u))**(3/2)
rho_2 = rho_2.to(u.g * u.cm**-3)

P_I = rho_2 * k_B * T / (mu * m_u)
P_I = P_I.to(u.g * u.cm**-1 * u.s**-2)
P_e = K_NR * (rho_2 / (mu_e))**(5/3)
P_e = P_e.to(u.g * u.cm**-1 * u.s**-2)
P_2 = P_I + P_e

print(rho_2, rho_c1)
print(P_2, P_c1)

M_2 = M_1 * (rho_2/rho_c1)**(-2) * (P_2/P_c1)**(3/2)
M_2 = M_2.to(u.M_sun)
print(M_2)
```
